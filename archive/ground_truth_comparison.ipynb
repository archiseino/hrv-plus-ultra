### Comparison between Ground Truth (BVP for each Task)

All ground truth dataset are sampled around 64 Hz

```python
## Import Dependencies
import numpy as np
import mediapipe as mp
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
from glob import glob
import scipy.signal as signal
```

```python
## Set the Sampling Rate
fs = 64

## Open the CSV Ground Truth Rest Task (T1) Csv
ground_truth_t1 = pd.read_csv('../Samples/Original/bvp_s53_T1.csv', header=None).values.flatten() # 64 Hz

## Open the CSV Ground Truth Speech Task (T2) Csv
ground_truth_t2 = pd.read_csv('../Samples/Original/bvp_s53_T2.csv', header=None).values.flatten() # 64 Hz

## Open the CSV Ground Truth Aritmetic Task (T3) Csv
ground_truth_t3 = pd.read_csv('../Samples/Original/bvp_s53_T3.csv', header=None).values.flatten() # 64 Hz
```

### Working with T1 dataset

```python
## Plot T1 GT
plt.figure(figsize=(20, 5)) 
plt.plot(ground_truth_t1, color='red')
plt.title('Ground Truth Signal')
plt.tight_layout()
plt.show()
```

```python
## Normalize the GT Signal
ground_truth_t1 = (ground_truth_t1 - np.mean(ground_truth_t1)) / np.std(ground_truth_t1)

## Finds Peak Signal
peaks, _ = signal.find_peaks(
    x=ground_truth_t1,
    prominence=0.5,
)

## Calculating the Heart Rate and Show the Result
heart_rate = 60 * len(peaks) / (len(ground_truth_t1) / fs)

plt.figure(figsize=(20, 5))
plt.plot(ground_truth_t1, color='black')
plt.plot(peaks, ground_truth_t1[peaks], 'x', color='red')
plt.title(f'Heart Rate: {heart_rate:.2f}')
plt.tight_layout()
plt.show()
```
